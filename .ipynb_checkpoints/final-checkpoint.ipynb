# Data preparation

We didn't have data on the population levels for each neighborhood. Therefore, in our individual analyses, we each merged our respective datasets with the population data for each neighborhood. By using the formula "index/population * 100", we calculated the index rate for each neighborhood. 

```python
import pandas as pd

# Load the dataset
## Occupation rate
data_occupation = "occupation_rate.csv"
occupation_rate = pd.read_csv(data_occupation)

## Enrollment rate
data_enrollment = "enrollment_rate.csv"
enrollment_rate = pd.read_csv(data_enrollment)

## Firearm sizure rate
data_firearm = "incident_rate.csv"
firearm_sizure_rate = pd.read_csv(data_firearm)

# Sort each dataset by "Neighborhood"
occupation_rate_sorted = occupation_rate.sort_values(by="Neighborhood")
enrollment_rate_sorted = enrollment_rate.sort_values(by="Neighborhood")
firearm_sizure_rate_sorted = firearm_sizure_rate.sort_values(by="Neighborhood")


# Merge the datasets on the "Neighborhood" column
merged_data = occupation_rate_sorted.merge(enrollment_rate_sorted, on="Neighborhood")
merged_data = merged_data.merge(firearm_sizure_rate_sorted, on="Neighborhood")
print(merged_data.head(5))
```

Since we require a single consolidated index, it's necessary to first merge these three datasets. 

We selected 'Total occupation rate', 'enrollment rate', and 'incident rate' to represent the rates for occupations, school enrollment, and firearm seizures, respectively.

```python
final_data = merged_data[['Neighborhood', 'Total occupation rate', 'enrollment rate', 'incident rate']]
print(final_data)
```

# Create final index


To create a final index that synthesizes the three indices (total occupation rate, enrollment rate, incident rate) in order to evaluate the best neighborhood in Pittsburgh, we need to decide on how to weigh each of these components in the final index. 

We think assigning equal weight to each index or choosing weights that reflect their importance to the overall quality of the neighborhood. Here is what we did:

* weight_occupation = 1/3 * total occupation rate
* weight_enrollment = 1/3 * total enrollment rate
* weight_safety = 1/3 * (1-total firearm seizure incident rate) 

Specifically:

* Total Occupation Rate: 
A high occupation rate suggests a robust employment environment, which is crucial for the economic stability of the residents. It indicates that more people in the neighborhood are engaged in meaningful work, which can lead to higher disposable income, lower crime rates, and generally improved societal conditions.

* Total Enrollment Rate: 
Similarly, a high enrollment rate in schools reflects the value placed on education within the neighborhood. It also implies a level of investment in future generations, as education is closely linked to long-term socio-economic benefits like better job opportunities and increased community engagement.

* Total Safety Rate:
The total firearm seizure incident rate typically represents negative aspects of a neighborhood, such as safety concerns and higher crime rates. By inverting this rate (1 - incident rate), the formula transforms a negative indicator into a positive one. This inversion means that lower incident rates (which are desirable for safety) result in higher scores, thus promoting neighborhoods with fewer firearm incidents as more desirable.

```python
# Weights for each component
weight_occupation = 1/3
weight_enrollment = 1/3
weight_incident = 1/3

# Normalize the incident rate because a lower rate is better
# Normalize to make all indices in the same direction (higher is better)
final_data['normalized_incident_rate'] = 1 - final_data['incident rate']

# Calculate the final index
final_data['final_index'] = (final_data['Total occupation rate'] * weight_occupation +
                             final_data['enrollment rate'] * weight_enrollment +
                             final_data['normalized_incident_rate'] * weight_incident)

# Sort data by the final index in descending order to find the best neighborhood
final_data_sorted = final_data.sort_values(by='final_index', ascending=False)

# Display the sorted DataFrame to see the neighborhoods ranked
print(final_data_sorted[['Neighborhood', 'final_index']])
```

# Result

Rank the best neiborhood based on final index (count for safety, education and socialeconomic status)

```python
import matplotlib.pyplot as plt

# Sorting the data by 'final_index' in descending order
final_data_sorted = final_data.sort_values(by='final_index', ascending=False)

# Create a bar chart
plt.figure(figsize=(12, 8))  
plt.barh(final_data_sorted['Neighborhood'][::-1], final_data_sorted['final_index'][::-1], color='skyblue')  # Reversing order to have the best on top
plt.xlabel('Final Index for Best neighbor') 
plt.ylabel('Neighborhood')  
plt.title('Neighborhood Ranking Based on Final Index')  
plt.grid(axis='x', linestyle='--', alpha=0.6)  
plt.tight_layout()

# Show the plot
plt.show()
```

```python
for index, row in final_data_sorted.head(20).iterrows():
    print(f"Neighborhood: {row['Neighborhood']}")
    print(f"Final Index for Best neighbor: {row['final_index']}")
    print()
```
