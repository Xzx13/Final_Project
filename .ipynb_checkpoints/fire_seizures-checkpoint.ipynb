# Analyzing Firearm Seizure Data in Pittsburgh
### Introduction
------------
The goal of this project is to identify the best neighborhood in Pittsburgh by analyzing firearm seizure data. 
The assumption is that fewer firearm seizures in a neighborhood indicate higher safety levels. 
This analysis might help in finding the best nrighborhood in the town when we analyze our 3 datasets together in the final presentation jupyter notebook.

### Approach
--------
We will begin by loading the data on firearm seizures, followed by data cleaning to prepare it for analysis. 
Descriptive statistics and visualizations will be used to compare neighborhoods based on the number of seizures.

### The Metric
----------
The primary metric for determining the 'best' neighborhood is the number of firearm seizures, 
where fewer seizures are assumed to indicate a safer neighborhood. This dataset provides details on firearm seizures 
across various neighborhoods in Pittsburgh.

----------
The dataset contains details about fire incidents in Pittsburgh, including fields such as incident type, address, the neighborhood involved, date/time of the alarm, and geographical coordinates. Here's a summary of the key fields in the dataset that can be relevant for the analysis:

incident_type and type_description: These fields describe the type of fire incident.
neighborhood: This is crucial as it indicates the neighborhood where the incident occurred.
alarm_time: The date and time when the fire alarm was reported.
latitude and longitude: Geographic coordinates of the incident location.

```python
# Importing Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Load the Data
data_path = 'firearm_seizures.csv'
firearm_data = pd.read_csv(data_path)

#Data Overview
print("Data Overview:")
print(firearm_data.head())
```

```python
# Rename 'neighborhood' to 'Neighborhood' and keep only this column
firearm_data_renamed = firearm_data.rename(columns={'neighborhood': 'Neighborhood'})

data_population = "population.csv"
population_data = pd.read_csv(data_population)
population_data['total population'] = population_data['Estimate; Total'] + population_data['Margin of Error; Total']

# Merge the two dataframes on the "Neighborhood" column
firearm_population = pd.merge(firearm_data_renamed, population_data, on="Neighborhood", how="inner")
firearm = firearm_population[['Neighborhood', 'total population']]
#Remove NAs
firearm = firearm.dropna()
#Counts the firearm incidents
seizures_per_neighborhood = firearm['Neighborhood'].value_counts().reset_index()
seizures_per_neighborhood.columns = ['Neighborhood', 'incidents']

firearm_final = pd.merge(seizures_per_neighborhood, firearm, on="Neighborhood", how="left")

# Create a new column for incident rate
firearm_final['incident rate'] = firearm_final['incidents'] / firearm_final['total population']
firearm_final = firearm_final.drop_duplicates()

print(firearm_final.head())
```

```python
lowest_firearm_final = firearm_final.sort_values(by='incidents', ascending=True)

plt.figure(figsize=(12, 8))

# Correctly reference the data for plotting
# 'Neighborhood' should be used for the x-axis and 'incidents' for the y-axis
sns.barplot(x='Neighborhood', y='incidents', data=lowest_firearm_final)

# Adding the title and labels with appropriate adjustments
plt.title('Firearm Seizures Per Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Number of Firearm Seizures')

# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=90)

plt.show()
```

```python
# Sort the data by 'incident rate' in descending order and select the top 5
lowest_5_firearm_final = firearm_final.sort_values(by='incident rate', ascending=True).head(5)

plt.figure(figsize=(12, 8))

# Correctly reference the sorted and limited data for plotting
# 'Neighborhood' for the x-axis and 'incident rate' for the y-axis
sns.barplot(x='Neighborhood', y='incident rate', data=lowest_5_firearm_final)

# Adding the title and labels with appropriate adjustments
plt.title('5 Neighborhoods with Lowest Firearm Seizure Rate')
plt.xlabel('Neighborhood')
plt.ylabel('Firearm Seizures Rate')

# Rotate the x-axis labels to prevent overlap
plt.xticks(rotation=45)  # Reduced rotation as fewer items

plt.show()
```

```python
firearm_final_sort = firearm_final.sort_values(by='incident rate', ascending=False)
output_path = 'Final_Project/incident_rate.csv'

# Save the sorted dataframe to a CSV file
firearm_final_sort.to_csv(output_path, index=False)
```
